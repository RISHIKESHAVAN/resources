```python
# Lists and Tuples:
# These are called compound data types and are one of the key types of data structures in Python.

# Tuples:
# Tuples are an ordered sequence. They are written as comma separated elements within paranthesis.
# All datatypes can be stored within a tuple. The datatype of the variable is tuple.
tuple1 = ("All", 10, 5.65)
type(tuple1) # tuple

# Each element of a tuple can be accessed via an index that starts from 0.
tuple1[0] # 'All'
# In Python, we can use negative index.
tuple1[-1] # 5.65

# We can concatenate or combine tuples by adding them.
tuple2 = tuple1 + ("Hard Rock", 15)
tuple2 # ('All', 10, 5.65, 'Hard Rock', 15)
# Tuples can be sliced. The beginning index is included and the ending index is excluded.
tuple2[0:3] # ('All', 10, 5.65)
tuple2[3:5] # ('Hard Rock', 15)

# Tuples are immutable. You can't modify the values in it. But you can assign it a different value. To understand why tuples are
# immutable:
ratings = (1, 59, 63, 57) # ratings is a new tuple
ratings1 = ratings # 'ratings1' is NOT a separate new variable with a separate value of (1, 59, 63, 57). It is just a reference for
                   # the same (1, 59, 63, 57) of 'ratings'. Thus if the value of 'ratings' is modified, then 'ratings1' is also
                   # affected. Thus tuples are immutable to prevent this.
# But ratings can be assigned a new value.
ratings = (5,10) # Now, ratings: (5,10) and ratings1: (1, 59, 63, 57)
# As a consequence of immutability, if we would like to manipulate a tuple, we must create a new tuple instead.
# For example, if we would like to sort a tuple, we use the function sorted.
ratings1Sorted = sorted(ratings1) # The input is the original tuple. The output is a new sorted tuple.

# A tuple can contain other tuples as well as other complex data types; this is called nesting.
# The elements inside a nested tuple can be accessed via the standard indexing methods.
NT = (1, 2, ("pop", "rock"), (3, 4),("disco", (5,6)))
NT[2] # ('pop', 'rock')
NT[2][0] # 'pop'
NT[4][1][0] # 5
```

```python
# Lists:
# Lists are also an ordered sequence. A list is represented with square brackets.
# In many respects lists are like tuples, one key difference is they are mutable.
L = ["Michael Jackson", 10, 12.56]
# Lists can contain strings, floats, integers. We can nest other lists.
# We also nest tuples and other data structures; the same indexing conventions apply for nesting.
# Like tuples, each element of a list can be accessed via an index.
# Charasteristics similar to tuples: -ve indexing, slicing, concantenation.
# But Lists are mutable.
L.extend(["pop",10]) # This adds the params as 2 elements to the existing list. It does NOT create a separate list.
L # ['Michael Jackson', 10, 12.56, 'pop', 10]
# append() and extend() are similar. But the key difference is, append() adds the params as one elements.
L.append(["pop", 15])
L # ['Michael Jackson', 10, 12.56, 'pop', 10, ['pop', 15]]
L[2] = 15.62
L # ['Michael Jackson', 10, 15.62, 'pop', 10, ['pop', 15]]
del(L[5]) # deletes the element at the specified index.
L # ['Michael Jackson', 10, 15.62, 'pop', 10]
# We can convert a string to a list using split. split() converts every group of characters separated by a space into 
# an element of a list.
"hard rock".split() # ['hard', 'rock']
# We can use the split function to separate strings on a specific character, known as a delimiter.
"A,B,C,D".split(",") # ['A', 'B', 'C', 'D']

# When we set one variable, B equal to A, both A and B are referencing the same list.
# Multiple names referring to the same object is known as aliasing.
A = [1,2,3]
B = A
# Now, both B and A are referencing to the same list object. So when you change the value of A, is affects B too.
A[0] = 10
B[0] # 10
# however, you can clone list “A” by using the following syntax.
B = A[:]
# Now, "A" references one list and “B” references a new copy or clone of the original list.
# Now if you change “A”, "B" will not change.
A[0]=1
B[0] # 10
A[1:] # This return all values after the index 1.
```
