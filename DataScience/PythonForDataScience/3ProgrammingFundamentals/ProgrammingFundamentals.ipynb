```python
# Conditions
# Comparison operations compare some value or operand, then, based on some condition they produce a Boolean.

# Equality test
a = 6
a == 7 # False
# Inequality test
a > 6 # False
a >= 6 # True
a < 6 # False
a <= 6 # True
a != 6 # False
"A" == "B" # False
"A" != "B" # True
```

```python
# Branching:
# Branching allows us to run different statements for a different input.

# if statement
age = 17
if (age>18): # Brackets are not necessary
    print("You can enter.") # Block to be executed if the condition was True.
print("Move on.") # Always printed.
# if-else statement
age = 17
if (age>18):
    print("You can enter AC/DC concert.")
else:
    print("You can enter Meatloaf concert.")
print("Move on.") # Always printed.
# elif statement: short for “else if,” allows us to check additional conditions, if the proceeding condition is false.
age = 18
if (age>18):
    print("You can enter AC/DC concert.")
elif (age==18):
    print("Go see Pink Floyd")
else:
    print("Go to Meatloaf concert.")
print("Move on.")
```

```python
# Logic operators:
# Logic operations take Boolean values and produce different Boolean values.

# not() operator produces the opposite
not(True) # False
# "or" takes in two Boolean values and produces a new Boolean value.
# +-------+-------+--------+
# |   A   |   B   | A or B |
# +-------+-------+--------+
# |  True | True  |  True  |
# +-------+-------+--------+
# | True  | False |  True  |
# +-------+-------+--------+
# | False | True  |  True  |
# +-------+-------+--------+
# | False | False | False  |
# +-------+-------+--------+
(5<6) or (7<6) # True
# "and" takes in two Boolean values and produces a new Boolean value.
# +-------+-------+---------+
# |   A   |   B   | A and B |
# +-------+-------+---------+
# |  True | True  |  True   |
# +-------+-------+---------+
# | True  | False |  False  |
# +-------+-------+---------+
# | False | True  |  False  |
# +-------+-------+---------+
# | False | False |  False  |
# +-------+-------+---------+
(5<6) and (7<6) # False
```

```python
# Loops:
# The range function outputs an ordered sequence as a list “i”.  Please note: if you use Python 3 the range() will not 
# generate a list explicitly like in Python 2.
range(3) # [0,1,2]
range(10,15)

# for loops: (We will focus on lists. But many of the procedures can be used on Tuples)
# Loops perform a task over and over.
squares = ["red","yellow","blue","orange"]
for i in range(0,4): # for i in [0,1,2,3], the value of i is incremented for each iteration.
    squares[i] = "white"
squares
# We can also iterate through a list or tuple directly in Python, without using indexes.
squaresNew = ["red","yellow","blue","orange"]
for square in squaresNew:
    square #red \n yellow \n blue \n orange
# enumerate() is a useful function for iterating data. It can be used to obtain the index and the element in the list. 
for i,square in enumerate(squaresNew):
    print(str(i) + square)
    
# while loops:
# Instead of executing a statement a set number of times, a while loop will only run if a condition is met.
squares1 = ["Orange","Orange","Orange","Purple","Blue"]
newSquares1 = [] #  the list is of indeterminate size.
i = 0
while (squares1[i] == "Orange"):
    newSquares1.append(squares1[i])
    i = i + 1
newSquares1
```

```python
# Functions:
# Functions take some input then produce some output or change.
# Python has many built-in functions.

# Built-in functions in Python:
# len() takes in an input of type sequence, such as a string or list, or type collection, such as a dictionary or set, and
# returns the length of that sequence or collection.
len([0,59,2,78,5]) # 5
# sum() takes in an iterable like a tuple or list and returns the total of all the elements.
sum([0,59,2,78,5])
# There are two ways to sort a list. The function sorted() and the method sort()
# sorted() is a function, meaning it will produce a new output.
a = [0,59,2,78,5]
aSorted = sorted(a) # So, this produces a new list with the sorted elements. The order of elements in list 'a' is unchanged.
# However, the method sort does not produce a new object as an output. Rather it changes the order of elements in 'a'.
a.sort() 
a # [0, 2, 5, 59, 78]
```

```python
# Building our own functions:
# To define a function, we start with the keyword 'def'. The name of the function should be descriptive of what it does. 
# We have the function formal parameter in the parentheses. Followed by a colon.
# We have a code block with an indent and return or output the final value. 
def addition(a):
    return a + 1

# After we define the function, we can call it. 
addition(5) # 6

# It's customary to document the function on the first few lines; this tells anyone who uses the function what it does. 
# This documentation is surrounded in triple quotes - """. 
def addition(a):
    """
    adds 1 to a
    """
    return a + 1
help(addition) # this returns the documentation of the function.

# A function can have multiple parameters. The parameters can be of any datatype that allow the computations of the method.
# This property will make coding simpler, but you must test your code more thoroughly. 
def mult(a,b):
    return a * b
mult(1,6) # 6
mult(2," HI") # ' HI HI'

# In many cases a function does not have a return statement. In these cases, Python will return the special “None” object.
def pr():
    print("Hi")
pr() # Hi

# Python doesn’t allow a function to have an empty body, so we can use the keyword 'pass', which doesn’t do anything,
# but satisfies the requirement of a non-empty body.
def noWork():
    pass
print(noWork()) # None
# In the background, if the return statement is not called, Python will automatically return a None.

# Loops can be used inside functions

# Variadic parameters allow us to input a variable number of elements.
def artistNames(*names):
    for name in names:
        print(name)
artistNames("Michael Jackson","AC/DC","Pink FLoyd")
artistNames("Michael Jackson","AC/DC")
```

```python
# Scope of a variable:
# The scope of a variable is the part of the program where that variable is accessible. 
# Variables that are defined outside of any function are said to be within the global scope, meaning they can be accessed 
# anywhere after they are defined.

def addDC(y):
    x = y+"DC" # scope of this x is within the function. Local variables only exist within the scope of a function.
    print(x) # this is the x within the function.
    return x # After the value is returned, the scope of the function is deleted.
# if there is any variable used within the function that is not defined within the function, then Python checks the
# Global scope.
x = "AC" # this is a global variable. it is accessible anywhere after it is defined.
z = addDC(x) # When we call the function, we enter a new scope or the scope of AddDC. 
print(x) # this is the global x. value is still AC
print(z) # ACDC
# -> Variables inside the global scope can have the same name as variables in the local scope with no conflict.

# We can define a global variable from within a function.
def pop():
    global claimedSales
    claimedSales = 45
    return claimedSales
pop()
print(claimedSales) # 45
```

```python
# Objects:
# Each data type in python is an object.
# Every object has the following: 
#    > a type 
#    > internal representation, 
#    > a set of functions called "methods" to interact with the data. 
# An object is an instance of a particular type.
a = 1 # we are creating an instance of type integer, or we are creating an integer object.
```

```python
# Methods:
# A class or type’s methods are functions that every instance of that class or type provides.
# It’s how you interact with the object.
Ratings = [1,59,36,74,95,8,3]
Ratings.sort() # sort() is the method
# Methods change the state of the objects.
```

```python
# Class:
# You can create your own type or class in Python.
# The class has data attributes and methods.
# We then create an instances or instances of that class or objects.
# To create a class, you need:
#    > to use the "class" keyword
#    > the class name
#    > to mention the parent class in the ()
# Once we define our class, we then initialize each instance of the class with data attributes using the class constructor.

class Circle(object):
    def __init__(self, radius,colour): # Constructor. "self" refers to the newly created instance of the class.
        self.radius = radius # sets the input value to the radius of the instance.
        self.colour = colour
# After we have created the class, we can create an object of class circle using the constructor.
redCircle = Circle(10,"Red") # This created
redCircle.colour # Red
redCircle.radius = 5 # this changes the radius attribute of the object directly.
```
