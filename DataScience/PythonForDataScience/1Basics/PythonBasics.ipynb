```python
# String Operations:
# A string is a sequence of characters. single / double quotes.
# Each element in the sequence can be accessed using an index represented by the array of numbers.
Name= "Michael Jackson"                       # Index	=> 0	1	2	3	4	5	6
Name[0] # M
Name[6] # l
# We can also use negative indexing with strings.
#  Index   => -7	-6	-5	-4	-3	-2	-1
Name[-7] # J
Name[-1] # n
# We can think of string as a list or tuple. We can treat the string as a sequence and perform sequence operations.
Name[0:4] # Mich
Name[8:12] # Jack
# We can also input a stride value as follows. 
Name[::2] # "McalJcsn" - The 2 indicates we select every second variable.
# We can also incorporate slicing. 
Name[0:4:2] # "Mca" - This returns every second value up to index four.
# We can use the “len” command to obtain the length of the string. 
len(Name)
# We can concatenate or combine strings using + symbol.
Name + " is the best."
# Replicate values of a string
3 * Name
# You cannot change the value of the string, but you can create a new string. Strings are immutable.
#Name[0] = "J" # throws error - 'str' object does not support item assignment
Name = Name + " is the best."
Name
# Backslashes represent the beginning of escape sequences.
print("Michael Jackson \n is the best.") # \n represents a new line
print("Michael Jackson \t is the best.") # \t represents a tab
print("Michael Jackson \\ is the best.") # \\ places a \
print(r"Michael Jackson \ is the best.") # same as above
```
